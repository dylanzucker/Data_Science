### 5. Read the reference for KMeans: https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html#sklearn.cluster.KMeans

### Now, create a clustering using KMeans. Let's assume you have prior knowledge that there are indeed 5 clusters, but you aren't aware of ground truth. Thus, you can only assess the SSE and silhouette coefficient for your clustering.
### Print the inertia_ (SSE) and the silhouette_score results for this clustering.

```python
kmeans = KMeans(n_clusters=5, random_state=0).fit(df_data[["x","y"]])
print(kmeans.inertia_)
y_pred = kmeans.predict(df_data[["x","y"]])
print(silhouette_score(df_data[["x","y"]], y_pred))
```

### 6. Generate a scatterplot again, however, now color the points based off of the labels generated by the clustering. Add the centroids using a distinguishing color and size, such as black with an alpha=0.5.


```python
#fig = plt.figure(figsize=(15, 10))
plt.scatter(df_data["x"], df_data["y"], c = y_pred, cmap="viridis")
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c="black", alpha = 0.5)
plt.show()
```

### 7. Let's take a step backward and assume that you aren't even sure about how many clusters you have. Evaluate cluster sizes from 2 through 10. However, create a data frame that stores the SSE and the silhouette score for each k. Show the resulting data frame, and show a line plot of both on separate graphs.

```python
inertias = []
sils = []
ks = range(2,11)
for k in ks:
    # Create a KMeans instance with the number of clusters k
    model = KMeans(n_clusters = k, init = 'k-means++' )
    # Fit model and obtain cluster labels
    y_pred = model.fit_predict(df_data[["x","y"]])
    inertias.append(model.inertia_)
    sils.append(silhouette_score(df_data[["x","y"]], y_pred))
```

```python
fig = plt.figure(figsize=(20, 10))
plt.plot(ks, inertias, "-o")
plt.xlabel("Number of clusters, k")
plt.ylabel("Inertia")
plt.xticks(ks)
plt.show()
```

```python
fig = plt.figure(figsize=(20, 10))
plt.plot(ks, sils, "-o")
plt.xlabel("Number of clusters, k")
plt.ylabel("Silhouettes")
plt.xticks(ks)
plt.show()
```

### 9. Re-generate your clustering for k=5. You will likely notice that your labels will be different colors than your original labels. That's annoying, but there's not much you can do about it unless you reassign your labels using a simple mapping.
### Let's bring in our ground truth. Write the code to remap your cluster labels to the same labels as your ground truth.
### (HINT: How? Perhaps the easiest approach is to use the contingency_matrix function in sklearn.metrics.cluster package. Show the output of that, then think about how you can use that to get the best mapping.)
### Once you have your cluster labels remapped, then plot your data, but keep every point that has been assigned the correct cluster label as black, and show those points that were not grouped into the correct cluster as red.


```python
def check_cm(cms):
    for row in cms:
        if (row==0).sum() < len(row) - 1:
            return False
    return True
```

```python
# Create a KMeans instance with the number of clusters k
model = KMeans(n_clusters = 5, init = 'k-means++' )
# Fit model and obtain cluster labels
y_pred = model.fit_predict(df_data[["x","y"]])
cm = contingency_matrix( df_data["target"], y_pred)
MAX_ITER = 500
i = 0
stop = check_cm(cm)
while i < MAX_ITER and stop == False:
    y_pred = model.fit_predict(df_data[["x","y"]])
    cm = contingency_matrix( df_data["target"], y_pred)
    stop = check_cm(cm)
    i += 1
print(cm)
print(i)
```

```python
df_data["correct"] = (y_pred != df_data["target"])
#fig = plt.figure(figsize=(15, 10))
plt.scatter(df_data["x"], df_data["y"], c = df_data["correct"], cmap="binary")
plt.show()
```

### 10. Since we do have ground truth, there are several metrics that are used to assess the quality of your clustering. Read the section on clustering performance evaluation in Scikit-learn's documentation page: https://scikit-learn.org/stable/modules/clustering.html#clustering-evaluation

### Pay close attention to the following: Adjusted Rand Index, Homogeneity, Completeness, and V-measure. In this section, very briefly describe each of these four measures. Though we did not explicitly cover them in class, these are good to know.

Adjusted Rand Index: Measures the similarity between the ground truth cluster assignments and the assigned cluster assignments.

Homogeneity: When only members of a single class are in each cluster

Completeness: Every single member from a specific class is assigned to the same cluster.

V-measure: The harmonic mean of the homogenity score and the completeness score

### 11. Create a new data frame that contains these four measures for all KMeans clusterings of k between 2-10. As before, show your data frame, then generate a plot for each. However, since each of these falls on the same scale, generate them on either one plot, or show them as subplots.

```python
df_metrics = pd.DataFrame(columns=["ARI", "homogeneity", "completeness", "v_meas"])
ks = range(2,11)
for k in ks:
    # Create a KMeans instance with the number of clusters k
    model = KMeans(n_clusters = k, init = 'k-means++' )
    # Fit model and obtain cluster labels
    y_pred = model.fit_predict(df_data[["x","y"]])
    homogeneity = metrics.homogeneity_score(df_data["target"], y_pred)
    ari = metrics.adjusted_rand_score(df_data["target"], y_pred)
    comp = metrics.completeness_score(df_data["target"], y_pred) 
    v_meas = metrics.v_measure_score(df_data["target"], y_pred)
    row = pd.DataFrame({"ARI": ari, "homogeneity": homogeneity, "completeness": comp, "v_meas":v_meas}, index = [k])
    df_metrics = df_metrics.append(row)
    
df_metrics
```
