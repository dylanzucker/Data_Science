### 1. Create a dataframe named df_uniform that contains 1000 observations. It should have two variables, named x and y. For each observation, x should be generated from a uniform distribution between 10 and 90, and y should be generated from a uniform between 20 and 80.

```python
np.random.seed(1234) # Set for redoability
df_uniform = pd.DataFrame([pd.Series(np.random.uniform(10, 90, 1000), name = "x"), pd.Series(np.random.uniform(20, 80, 1000), name = "y")])
df_uniform = df_uniform.T
df_uniform.head(10)
```

### 12. What is a quantile? What is a quartile? What is an Inter-quartile range (IQR)? Interpret the boxplot results in these terms.

Quantiles are points that divide the range of a probability distribution into intervals with equal probabilities. Quartiles are a type of quantile, where the first quartile is the midpoint between the minimum value and the median of the data (the 25th percentile). The second quartile is the median. The third quartile is the midpoint of the median and the maximum value of the data (the 75th percentile). The IQR is the third quartile - the first quartile. The Normal plot shows the x values have a Q1 = 40, Q2 = 50, Q3 = 60, and IQR of 20, and the y values have a Q1 of about 50, Q2 of about 52, Q3 of about 55, and an IQR of 5.   

### 13. Read about the quantile() method for data frames, and use it to numerically show the 25th, median, and 75th percentiles, and compute the IQR (Inter-quartile range) for both variables, on both data frames. Compare and contrast.

The IQR is much larger for the uniform distribution.

```python
print("25th, normal: \n", df_normal.quantile(.25))
print("median, normal: \n", df_normal.quantile(.50))
print("75th, normal: \n", df_normal.quantile(.75))
print("IQR, normal: \n", df_normal.quantile(.75) - df_normal.quantile(.25))

print("25th, uniform y: \n", df_uniform.quantile(.25))
print("median, uniform y: \n", df_uniform.quantile(.50))
print("75th, uniform y: \n", df_uniform.quantile(.75))
print("IQR, uniform: \n", df_uniform.quantile(.75) - df_uniform.quantile(.25))
```

### 14. What is the standard definition of an outlier, in terms of IQR? Use the IQR to determine what the outliers are for each variable in each dataset, if any.

Outliers are all points below the 25th percentile - 1.5*IQR and above 75th percentile + 1.5*IQR.

```python
uni_iqr = df_uniform.quantile(.75) - df_uniform.quantile(.25)
norm_iqr = df_normal.quantile(.75) - df_normal.quantile(.25)
print("normal outliers x: \n", df_normal[(df_normal["x"] < df_normal.x.quantile(.25) - 1.5 * norm_iqr["x"]) | (df_normal["x"] > df_normal.x.quantile(.75) + 1.5 * norm_iqr["x"])]["x"])
print("normal outliers y: \n", df_normal[(df_normal["y"] < df_normal.y.quantile(.25) - 1.5 * norm_iqr["y"]) | (df_normal["y"] > df_normal.y.quantile(.75) + 1.5 * norm_iqr["y"])]["y"])
print("uniform outliers x: \n", df_uniform[(df_uniform["x"] < df_uniform.x.quantile(.25) - 1.5 * uni_iqr["x"]) | (df_uniform["x"] > df_uniform.x.quantile(.75) + 1.5 * uni_iqr["x"])]["x"])
print("uniform outliers y: \n", df_uniform[(df_uniform["y"] < df_uniform.y.quantile(.25) - 1.5 * uni_iqr["y"]) | (df_uniform["y"] > df_uniform.y.quantile(.75) + 1.5 * uni_iqr["y"])]["y"])
```

## Part III  Some basic data preprocessing

### 27) Create an additional variable in df_uniform called x_fac1 that represents a factor with 3 levels, "X1", "X2", and "X3". You should discretize according to equal width bins over the distribution of x. (Divide the range of x into three.)

```python
def split_x(df, x):
    mi=df["x"].min()
    ma=df["x"].max()
    range_x = ma - mi
    if x <= mi + (range_x * 1/3):
        return "X1"
    elif x <= mi + (range_x * 2/3):
        return "X2"
    return "X3"
df_uniform["x_fac1"] = df_uniform.x.apply(lambda x: split_x(df_uniform, x))
df_uniform.head(10)
```

### 28. Create an additional variable in df_uniform called x_fac2 that represents a factor with 3 levels, "X1", "X2", and "X3". This time, you should discretize using equal depth bins over the distribution of x. Select your division criteria such that there are an equal number of data in each bin. Verify that the distribution of your data each has the same number of data (within 1).

```python
def split_x_depth(df, x):
    q1=df["x"].quantile(1/3)
    q2=df["x"].quantile(2/3)
    if x <= q1:
        return "X1"
    elif x <= q2:
        return "X2"
    return "X3"
df_uniform["x_fac2"] = df_uniform.x.apply(lambda x: split_x_depth(df_uniform, x))
df_uniform["x_fac2"].value_counts()
```

### 29. Create a side by side scatter plot showing the distribution of df_uniform, using x_fac1 as the color for one plot, and x_fac2 as the color for your other plot.

```python
fig, axs = plt.subplots(ncols=2)
sns.scatterplot(x="x", y="y", hue="x_fac1", data=df_uniform, marker="v", ax=axs[0])
sns.scatterplot(x="x", y="y", hue="x_fac2", data=df_uniform, marker="4", ax=axs[1])
```
