### 15. Read in the Chipotle dataset:

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chip_df = pd.read_csv(url, sep = '\t')
```

### 16. Recall that this dataset is a set of transactions, where each observation represents one item purchased as part of an order_id. However, the data are not read in this way. You need to transform this dataset to a collection of binary encoded transactions, where each row represents ONE transaction, and the columns are binary encoded variables, with each variable representing ONE item available for purchase at Chipotle. Convert your data. Your resulting data frame should have an index representing the order_id, and columns representing each possible item from the item_name variable. For now, a transaction will ignore the quantity of item purchased.

```python
bin_chip_df = pd.get_dummies(chip_df, columns=["item_name"]).groupby("order_id").sum().apply(lambda y: y.apply(lambda x : x if x == 0 else 1), axis = 1)
bin_chip_df.drop("quantity", axis=1, inplace=True)
bin_chip_df.head()
```

### 17. Too often, many start by considering a minimum support that is arbitrarily large. Go ahead and use the apriori method to generate frequent itemsets with a minsup value of 0.5. What happened?

There were none with a minsup of 0.5

```python
ap_chip = apriori(bin_chip_df, min_support=0.5, use_colnames=True)
ap_chip.head()
```

### 18. Take a step back. Your previous outcome is why you ALWAYS perform essential EDA tasks before you dive into mining a dataset! Report a table that shows the number of transactions each item occurred in, sorted in order of most frequent to least. NOTE: That number essentially represents the absolute support for 1- itemsets! So, include a column that shows the relative support (i.e. the fraction of total transactions.)

```python
x = bin_chip_df.apply(lambda x: x.sum()) 
table = pd.concat([x, x/bin_chip_df.shape[0]], axis = 1).sort_values(by=[0], ascending = False)
table.columns = ["support", "abs support"]
table.head()
```

### 19. Now, make a smarter decision. Like many large, real-world transaction datasets, data is sparse! You have many variables, and most observations use only a handful of them. This is the definition of a sparse dataset. You need a better minsup value. Regenerate frequent itemsets, but now use a minsup of 0.005. How many frequent itemsets were reported? Report your frequent items sorted by decreasing support order.

```python
ap_chip = apriori(bin_chip_df, min_support=0.005, use_colnames=True)
ap_chip.head()
```

### 20. In the context of association rules, explain the difference between support, confidence, lift, leverage and conviction.

The support is a measure of frequencey for an item set. Confidence is the degree of cettainty of that an item set will appear. The lift compares the actual confidence to the expected confidence. So lift is how much better the rule is at predicting than just guessing based on the expected value. The conviction is the frequencey that a rule will make the wrong prediction. The leverage is the difference of the observed frequency of two items appearing in the same set and the expected frequencey if they were indepedent.

### 25. These data suggest that the "Chicken Bowl" is the single most frequent item purchased for this particular location. Great. Let's dive into those chicken bowls more. As you know, Chipotle lets you customize your items. That's what the choice_description field is for.

### Process all of the items listed in the "choice_description" field by creating a new transaction dataset representing binary encoded transaction data for only Chicken Bowl. Note – this is tricky because the choice_description variable is read in as a long string. The string itself represents a list, and quite often, it's a list of lists. You need to process this to be an actual flattened list of items. Then, you can easily convert these to transactions.

```python
chick_df = chip_df[chip_df["item_name"].str.contains("Chicken Bowl", regex=False)]
chick_df["choice_description"] = chick_df["choice_description"].str.split(",", expand = True)
chick_df.dropna(axis=0, how = "any", inplace = True)

bin_chick_df = pd.get_dummies(chick_df, columns= ["choice_description"]).groupby("order_id").sum().apply(lambda y: y.apply(lambda x : x if x == 0 else 1), axis = 1)
bin_chick_df = bin_chick_df.drop("quantity", axis = 1)
bin_chick_df.head()
```
